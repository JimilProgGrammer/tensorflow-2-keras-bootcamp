### Part 2: Feature Engineering

```python
df.head()
```

```python
# Drop ID column - not a useful feature for prediction
df = df.drop('id',axis=1)
```

```python
# Convert string dates to datetime objects
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='month',y='price',data=df)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
# Drop the date
df = df.drop('date',axis=1)
```

```python
# Drop the zipcode
df = df.drop('zipcode',axis=1)
```

```python
# You coud further engineer this feature by categorizing value 0 as not renovated vs any year as renovated.
# However, you can leave it as is because higher year of renovation (more recent) has a high correlation with the sale price.
# So year 0 will have lowest price, and will go on increasing till the highest year in the data.
df['yr_renovated'].value_counts()
```

### Part 3: Preprocessing and Model Creation

```python
# Separate features and labels
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
# Perform split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Data Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model Creation: Imports
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Model Creation: Compiling
model = Sequential()

# Number of neurons for the layers can be determined from the number of features in your dataset
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
# Model Creation: Training
model.fit(x=X_train, y=y_train, validation_data=(X_test,y_test), batch_size=128, epochs=400)
```

### Part 4: Model Evaluation and Prediction

```python
losses = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(10,5))
losses.plot()
```

<p>In the plot above, if you see a spike in the val_loss (loss on validation data) after some epochs, this means that model is overfitting on the validation/test data. </p>

```python
# Model evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
predictions = model.predict(X_test)
print("MAE: " + str(mean_absolute_error(y_test, predictions)))
print("MSE: " + str(mean_squared_error(y_test, predictions)))
print("RMSE: " + str(mean_squared_error(y_test, predictions)**0.5))
print("Explained Variance Score: " + str(explained_variance_score(y_test, predictions)))
```

```python
plt.figure(figsize=(12,8))
error_df = pd.DataFrame(list(zip(y_test, predictions)),columns=['Y Test','Predictions'])
sns.scatterplot(x='Y Test', y='Predictions', data=error_df)
plt.plot(y_test, y_test, 'mediumseagreen')
```

<p>The green line is the perfect prediction line. Above plot indicates that we are heavily punished for predicting prices of the outliers i.e. houses with extremely high prices. We can continue to train our model for more epochs since we are not overfitting or we can train the model over the bottom 99% and leave the top 1% outliers out.</p>
