### Part 1: Preparing the data

```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python
# Perform train-test split
from sklearn.model_selection import train_test_split
X = df[['feature1','feature2']].values
y = df['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Normalizing data to prevent exploding and vanishing gradient problem
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Fit the scaler only on the training set to prevent data leakage from the test set 
# (We don't have prior information on the test set and we don't want to peek into it)
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Part 2: Creating and Training the model

```python
# Imports
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Building the model
model = Sequential()

# Add layers individually
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='rmsprop', loss='mse')
```

```python
# Train model
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
# Plotting training history
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

### Part 3: Model Evaluation

```python
# First method: Returns the loss metric (MSE in this case) from model after evaluation
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_train, y_train, verbose=0)
```

```python
# Second method: Get RMSE, MSE and MAE
y_pred = pd.Series(model.predict(X_test).reshape(300,))
pred_df = pd.DataFrame(list(zip(y_test, y_pred)),columns=['Y Test','Y Predictions'])
pred_df.head()
```

```python
# Plot the two against each other
sns.scatterplot(x='Y Test',y='Y Predictions',data=pred_df)
```

```python
import math
from sklearn.metrics import mean_absolute_error, mean_squared_error
print("MAE: " + str(mean_absolute_error(pred_df['Y Test'],pred_df['Y Predictions'])))
print("MSE: " + str(mean_squared_error(pred_df['Y Test'],pred_df['Y Predictions'])))
print("RMSE: " + str(math.sqrt(mean_squared_error(pred_df['Y Test'],pred_df['Y Predictions']))))
```

```python
# Predicting on brand new data
new_gem = [[998,1000]]
# NOTE: Always remeber to scale your features if your model is trained on scaled features
new_gem = scaler.transform(new_gem)
model.predict(new_gem)
```

```python
# Saving your weights and biases and using them later
model.save('gem_regression_model.h5')
from tensorflow.keras.models import load_model
loaded_model = load_model('gem_regression_model.h5')
# Use it like the previous model
loaded_model.predict(new_gem)
```
