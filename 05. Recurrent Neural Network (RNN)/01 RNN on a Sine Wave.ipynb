## Part 1: The Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,50,501)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
df = pd.DataFrame(data=y, index=x, columns=['Sine'])
df.head()
```

### Perform Train-Test Split

```python
test_percent = 0.1                 #10% test set
test_point = np.round(len(df)*test_percent)
```

```python
test_index = int(len(df)-test_point)
train = df.iloc[:test_index]
test = df.iloc[test_index:]
```

### Scale your Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(train)
scaled_test = scaler.transform(test)
```

## Part 2 - Batch Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
"""
Args for the TimeseriesGenerator object (converts the data into a time-stamped series)
:data: The X-value i.e. the timestamps
:targets: The Y-value i.e. the labels
:length: length of the output sequences
:batch_size: number of timeseries samples in each batch
"""
length = 25
batch_size = 1

generator = TimeseriesGenerator(data=scaled_train, targets=scaled_train, length=length, batch_size=batch_size)
```

```python
len(scaled_train)
```

```python
# len(generator) = len(scaled_train) - length_generator
#                =       451         -       2
len(generator)
```

```python
X,y = generator[0]
print("X = " + str(X) + "\nY = " + str(y))
```

```python
# What generator does is given the first two data points from the dataset i.e. 0.4999 and 0.5499
# it predicts the 3rd point in the sequence (y) which is nothing but 0.5993
scaled_train[0:(length+1)]
```

```python
# You can see that the first 25 points exhibit proper patterns of the sine wave that we want the network to learn
df[0.0:(0.0+25)].plot()
```

## Part 3 - Creating the Model

```python
length = 50
batch_size = 1
generator = TimeseriesGenerator(data=scaled_train, targets=scaled_train, length=length, batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
# A single X feature predicting the label Y
n_features = 1
```

```python
model = Sequential()
model.add(SimpleRNN(units=50, input_shape=(length, n_features)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator, epochs=5)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
first_eval_batch = scaled_train[-length:]
first_eval_batch = first_eval_batch.reshape((1,length,n_features))
model.predict(first_eval_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(scaled_test)):
    current_pred = model.predict(current_batch)[0]
    test_predictions.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
test['Predictions'] = true_predictions
test.head()
```

```python
test.plot(figsize=(12,8))
```

## Part 4 - Using LSTM instead of SimpleRNN

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
```

```python
# length of the training sequence for each data point in the batch must be less than
# the length of the validation sequence (testset in our case), if you want a validation generator 
length = 49

generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=1)
validation_generator = TimeseriesGenerator(data=scaled_test, targets=scaled_test, length=length, batch_size=1)
```

```python
model = Sequential()

model.add(LSTM(50, input_shape=(length,n_features)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator, epochs=20, validation_data=validation_generator, callbacks=[early_stopping])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
test_predictions = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(scaled_test)):
    current_pred = model.predict(current_batch)[0]
    test_predictions.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
test['LSTM_Predictions'] = true_predictions
test.plot(figsize=(12,8))
```

```python
full_scaler = MinMaxScaler()
scaled_data = full_scaler.fit_transform(df)

generator = TimeseriesGenerator(scaled_data, scaled_data, length, 1)

model = Sequential()
model.add(LSTM(50, input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit_generator(generator, epochs=4)
```

```python
forecast = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(25):
    current_pred = model.predict(current_batch)[0]
    forecast.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
forecast = full_scaler.inverse_transform(forecast)

# 50.1 is the new index of the unknown future
# 25 points are predicted at a space of 0.1
# so 50.1 + 2.5 = 52.6
forecast_index = np.arange(50.1,52.6,step=0.1)
```

```python
len(forecast)
```

```python
len(forecast_index)
```

```python
plt.plot(df.index, df['Sine'])
plt.plot(forecast_index, forecast)
```

```python
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['Sine'])
data = pd.concat([df,forecast_df])

plt.figure(figsize=(12,8))
plt.plot(data[:50.1].index, data[:50.1].values, label='Sine')
plt.plot(data[50.1:].index, data[50.1:].values, label='Prediction')
plt.legend()
```
