```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('..\DATA\RSCCASN.csv', parse_dates=True, index_col='DATE')
df.head()
```

```python
df.info()
```

```python
df.columns = ['Sales']
df.plot(figsize=(12,8))
```

```python
test_size = 18      # no of months
test_index = len(df) - test_size

train = df.iloc[:test_index]
test = df.iloc[test_index:]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

length = 12
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dense, LSTM
```

```python
n_features = 1

model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(length, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5)
validation_generator = TimeseriesGenerator(scaled_test, scaled_test, length=length, batch_size=1)

model.fit_generator(generator, epochs=20, validation_data=validation_generator, callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
test_prediction = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(test)):
    current_pred = model.predict(current_batch)[0]
    test_prediction.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_prediction)
test['Predictions'] = true_predictions
test.plot(figsize=(12,8))
```

```python
full_scaler = MinMaxScaler()
scaled_data = full_scaler.fit_transform(df)

length = 12
generator = TimeseriesGenerator(scaled_data, scaled_data, length=length, batch_size=1)

model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(length, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit_generator(generator, epochs=7)
```

```python
forecast = []
periods = 12

first_eval_batch = scaled_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(periods):
    current_pred = model.predict(current_batch)[0]
    forecast.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
forecast = scaler.inverse_transform(forecast)
forecast_index = pd.date_range(start='2019-11-01', periods=periods, freq='MS')
```

```python
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['Forecast'])
```

```python
ax = df.plot()
forecast_df.plot(ax=ax)
```

```python
final_data = pd.concat([df, forecast_df])
plt.figure(figsize=(12,8))
plt.plot(final_data[:335].index, final_data[:335].values, label='Sales')
plt.plot(final_data[335:].index, final_data[335:].values, label='Predictions')
plt.legend()
```
